```python
import csv
import datetime

# Menu.txt dosyasını oluşturma
with open('Menu.txt', 'w') as f:
    f.write('* Lütfen Bir Pizza Tabanı Seçiniz:\n')
    f.write('1: Klasik\n')
    f.write('2: Margarita\n')
    f.write('3: TürkPizza\n')
    f.write('4: Sade Pizza\n')
    f.write('* ve seçeceğiniz sos:\n')
    f.write('11: Zeytin\n')
    f.write('12: Mantarlar\n')
    f.write('13: Keçi Peyniri\n')
    f.write('14: Et\n')
    f.write('15: Soğan\n')
    f.write('16: Mısır\n')
    f.write('* Teşekkür ederiz!')

#Pizza üst sınıfını tanımlama
class Pizza:
    def __init__(self):
        self.description = "Unknown pizza"
        self.cost = 0.0
    
    def get_description(self):
        return self.description
    
    def get_cost(self):
        return self.cost
#Her pizza için alt sınıf oluşturma
class KlasikPizza(Pizza):
    def __init__(self):
        self.description = "Klasik Pizza"
        self.cost = 85.0

class MargheritaPizza(Pizza):
    def __init__(self):
        self.description = "Margherita Pizza"
        self.cost = 90.0

class TurkPizzasi(Pizza):
    def __init__(self):
        self.description = "Türk Pizzası"
        self.cost = 100.0

class DominosPizza(Pizza):
    def __init__(self):
        self.description = "Dominos Pizza"
        self.cost = 117.0
#Soslar için üst sınıf oluşturma
class Decorator(Pizza):
    def __init__(self, component):
        self.component = component

    def get_cost(self):
        return self.component.get_cost() + Pizza.get_cost(self)

    def get_description(self):
        return self.component.get_description() + ' ' + Pizza.get_description(self)
# Her sos için alt sınıf oluşturma
class Olive(Decorator):
    def __init__(self, component):
        self.description = "Zeytin"
        self.cost = 16.0
        Decorator.__init__(self, component)

class Mushroom(Decorator):
    def __init__(self, component):
        self.description = "Mantar"
        self.cost = 19.5
        Decorator.__init__(self, component)

class GoatCheese(Decorator):
    def __init__(self, component):
        self.description = "Keçi Peyniri"
        self.cost = 20.0
        Decorator.__init__(self, component)

class Meat(Decorator):
    def __init__(self, component):
        self.description = "Et"
        self.cost = 28.5
        Decorator.__init__(self, component)

class Onion(Decorator):
    def __init__(self, component):
        self.description = "Soğan"
        self.cost = 12.0
        Decorator.__init__(self, component)

class Corn(Decorator):
    def __init__(self, component):
        self.description = "Mısır"
        self.cost = 10.0
        Decorator.__init__(self, component)


#Ana(main) fonksiyonu açarak menüyü görüntüler ve pizzayı oluşturmaya başlar
def main():
    # Menüyü oynatma
    with open("Menu.txt", "r") as f:
        menu = f.read()
        print(menu)

    # Prompt user for pizza and toppings
    pizza = None

    #Bir pizza ve sos seçimi
    pizza_choice = int(input())

    if pizza_choice == 1:
        pizza = KlasikPizza()
    elif pizza_choice == 2:
        pizza = MargheritaPizza()
    elif pizza_choice == 3:
        pizza = TurkPizzasi()
    else:
        pizza = DominosPizza()

    print("Ve seçeceğiniz sos:\n11: Zeytin\n12: Mantarlar\n13: Keçi Peyniri\n14: Et\n15: Soğan\n16: Mısır")
    topping_choice = int(input())

    if topping_choice == 11:
        pizza = Olive(pizza)
    elif topping_choice == 12:
        pizza = Mushroom(pizza)
    elif topping_choice == 13:
        pizza = GoatCheese(pizza)
    elif topping_choice == 14:
        pizza = Meat(pizza)
    elif topping_choice == 15:
        pizza = Onion(pizza)
    elif topping_choice == 16:
        pizza = Corn(pizza)
    else:
        pass

  

    #fiyat hesaplama
    total_cost = pizza.get_cost()

    print("Toplam Fiyat: {} TL".format(total_cost))


    #Müşteri bilgilerini alma
    name = input("\nİsminiz ve Soyisminiz: ")
    tc_id = input("T.C. Kimlik numaranız: ")
    card_number = input("Kredi Kartı Numaranız: ")
    card_cvv = input("Kredi Kartı CVV numaranız: ")

    # Sipariş zamanının alınması
    def siparis_zamani():
      now = datetime.datetime.now()
      return now.strftime("%Y-%m-%d %H:%M:%S")


    # Siparişin Orders_Database.csv dosyasına kaydedilmesi
    with open("Orders_Database.csv", mode="a", newline="") as file:
        writer = csv.writer(file)
        writer.writerow([name, tc_id, card_number, pizza_choice, " with ", topping_choice, " sauce", siparis_zamani(), card_cvv])

    # Siparişin teyit edilmesi
    print("\nSiparişiniz için teşekkür ederiz!")
    print("Pizza seçiminiz: ", pizza_choice, " Sos seçiminiz: ", topping_choice, " Yakında size ulaşacaktır.")
    print("Toplam Ücret: $", total_cost)

main()    
```
